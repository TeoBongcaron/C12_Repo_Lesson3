```python
# Python Library Management System

# Install required libraries (This should already exist in colab)
!pip install requests

import pickle
import sqlite3
import requests
```

```python
# 1. Data Types and Variables
# Basic book details using different data types
book_title = "Python Programming"
book_quantity = 5
book_price = 29.99
is_available = True

print(f"Book: {book_title}, Quantity: {book_quantity}, Price: ${book_price}, Available: {is_available}")
```

```python
# 2. Lists
# Initial book list
books = ["Python Basics", "Advanced Python", "Data Science with Python"]

# Add two more books
books.append("Fluent Python")           # Adds to the end
books.insert(1, "Python Crash Course")  # Inserts at index 1

print("Updated book list:", books)
```

```python
# 3. Loops and Range()
print("\nAvailable books:")
# Print each book with its index
for i in range(len(books)):
    print(f"{i}: {books[i]}")
```

```python
# 4. User Input
new_book = input("Enter a new book title: ")
books.append(new_book)
print("Updated book list:", books)
```

```python
# 5. Functions and Conditional Statements
def check_availability(book_name):
    """
    Check if a book is available in the list.
    Returns True if found, False otherwise.
    """
    return book_name in books

# Test the function
print(f"Is 'Python Basics' available? {check_availability('Python Basics')}")
```

```python
# 6. File I/O
def save_books_to_file(filename):
    """
    Save the list of books to a text file.
    """
    try:
        with open(filename, 'w') as f:
            for book in books:
                f.write(f"{book}\n")
        print(f"Books saved to {filename}")
    except Exception as e:
        print("Error saving to file:", e)

def read_books_from_file(filename):
    """
    Read books from a text file and return as a list.
    """
    try:
        with open(filename, 'r') as f:
            return [line.strip() for line in f.readlines()]
    except FileNotFoundError:
        print(f"File {filename} not found.")
        return []
    except Exception as e:
        print("Error reading from file:", e)
        return []

# Test file I/O
save_books_to_file("books.txt")
print("Books read from file:", read_books_from_file("books.txt"))
```

```python
# 7. Pickle for Serialization
def save_books_pickle(filename):
    """
    Serialize and save the book list using pickle.
    """
    try:
        with open(filename, 'wb') as f:
            pickle.dump(books, f)
        print(f"Books pickled to {filename}")
    except Exception as e:
        print("Error saving pickle:", e)

def load_books_pickle(filename):
    """
    Load and deserialize the book list from a pickle file.
    """
    try:
        with open(filename, 'rb') as f:
            return pickle.load(f)
    except FileNotFoundError:
        print(f"Pickle file {filename} not found.")
        return []
    except Exception as e:
        print("Error loading pickle:", e)
        return []

# Test pickle functions
save_books_pickle("books.pkl")
print("Books loaded from pickle:", load_books_pickle("books.pkl"))
```

```python
# 8. SQLite Database Interaction
def create_books_table():
    """
    Create a SQLite table for storing books.
    """
    try:
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()
        cursor.execute('''CREATE TABLE IF NOT EXISTS books
                          (id INTEGER PRIMARY KEY AUTOINCREMENT,
                           title TEXT NOT NULL,
                           quantity INTEGER NOT NULL)''')
        conn.commit()
        conn.close()
        print("Books table created.")
    except Exception as e:
        print("Error creating table:", e)

def add_book_to_db(title, quantity):
    """
    Add a book to the SQLite database.
    """
    try:
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()
        cursor.execute("INSERT INTO books (title, quantity) VALUES (?, ?)", (title, quantity))
        conn.commit()
        conn.close()
        print(f"Book '{title}' added to database with quantity {quantity}.")
    except Exception as e:
        print("Error adding book to database:", e)

# Test database functions
create_books_table()
add_book_to_db("Python for Beginners", 10)
```

```python
# 9. Web API Interaction (GitHub API)

def get_python_repos():
    url = "https://api.github.com/search/repositories"
    params = {"q": "language:python", "sort": "stars", "order": "desc"}
    response = requests.get(url, params=params)
    data = response.json()
    return [repo['name'] for repo in data['items'][:5]]

# Test API function
print("Top 5 Python repositories on GitHub:", get_python_repos())

print("\nLibrary Management System operations completed.")


```
